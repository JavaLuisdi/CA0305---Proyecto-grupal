```python
import numpy as np
def broadcast_ops(X: np.ndarray, b: np.ndarray, w: np.ndarray) -> np.ndarray:
    
  """
La función lo que hará es sumar un vector fila y se aplica en cada fila de la 
matriz de manera que a la primera entrada del vector que se suma se aplica en la
primera entrada de la fila y así sucesivamente con todas las filas y después de 
ese resultado, se va a multiplicar un vector columna que la primera entrada del 
vector columna va a multiplicar la primera entrada de todas las columnas y así 
sucesivamente con todas las entradas de las columnas.

Variables
----------
X: es la matriz a modificar de dimensión (N,D)
b: es un vector que se sumará entrada por entrada de tamaño (D,)
w: es un vector que se va a multiplicar entrada por entrada de dimensión (W,)
Retorna
---------- 
Y = (X+b)*w 
  """
# hay que modificar el w para que tenga sentido 
  Y = (X + b) * w[:,None]

  return Y
```

### 2. Matrix Multiplication (Naive vs Vectorized)

```python
import numpy as np
def matmul_naive(A: np.ndarray, B: np.ndarray) -> np.ndarray:
  """
  La función va a resolver la multiplicación de matrices de una forma clásica
  es decir a través de que  se define la entrada Cij como la suma de los 
  productos de la entrada ik de la matriz A por la entrada kj de la matriz B
  sobre todos los k y así se tiene iterar para cada i y para cada j.
  Entonces vamos a hacer tres ciclos uno para mover el índice i otro para el 
  índice j y el último para que haga lo anteriormente descrito de la entrada Cij
  
  Variables
  ----------
  A: una matriz de tamaño N x M
  B: una matriz de tamaño R x S

  Retorna
  ----------
  C: La matriz resultante de hacer el producto matricial de A con B
  """
  # Verificar el tamaño de las matrices

  # Cantidad de filas de la matriz A
  rowA = A.shape[0]
  
  # Cantidad de columnas de la matriz A
  colA= A.shape[1]  
  
  # Cantidad de filas de la matriz B 
  rowB= B.shape[0]

  # Cantidad de columnas de la matriz B
  colB = B.shape[1]

  if colA == rowB:

  # Se crea la matriz C
    C = np.zeros((rowA,colB))

    # Iteramos sobre cada fila de A
    for i in range(rowA):
      # Itermaos sobre la cantidad de columnas de B
        for j in range(colB):
           # Iteramos para los k i.e. donde colA = rowB
            for k in range(colA):
              C[i,j] += A[i,k]* B[k,j]

    return C
  else:
     print("No se puede multiplicar estas matrices")
```

```python
import numpy as np
def matmul_vectorized(A: np.ndarray, B: np.ndarray) -> np.ndarray:
  """
  En numpy hay unas operaciones que hacen ese trabajo sucio que hicimos antes
  en un símbolo entonces vamos a establecerla

 Variables
 ----------
 A: una matriz de tamaño N x M
 B: una matriz de tamaño R x S

 Retorna
 ----------
  C: La matriz resultante de hacer el producto matricial de A con B
  """
  C = A @ B
  return C
```

### 4. Tensor Reshaping & Transposing

```python
import numpy as np
def reshape_and_transpose(x, B, C, H, W):
    """
    Convierte un vector 1D de imágenes desde formato aplanado
    al formato NHWC.

    Primero organiza los datos como NCHW: (B, C, H, W).
    Luego transpone los ejes para obtener NHWC: (B, H, W, C).

    Parámetros:
        x: vector 1D con B*C*H*W elementos.
        B: cantidad de imágenes del lote.
        C: cantidad de canales.
        H: altura de cada imagen.
        W: ancho de cada imagen.

    Retorna:
        np.ndarray con forma (B, H, W, C).
    """
    # 1. Verificar tamaño correcto
    if len(x) != B * C * H * W:
        raise ValueError("El tamaño de x no coincide con B*C*H*W")

    # 2. Convertir a arreglo NumPy
    x = np.array(x)

    # 3. Remodelar a formato NCHW: (B, C, H, W)
    tensor_nchw = x.reshape(B, C, H, W)

    # 4. Transponer a formato NHWC: (B, H, W, C)
    tensor_nhwc = np.transpose(tensor_nchw, (0, 2, 3, 1))

    return tensor_nhwc

    
#Prueba

x = list(range(1, 25))
B = 1
C = 2
H = 3
W = 4

resultado = reshape_and_transpose(x, B, C, H, W)

print(resultado)
print("Forma final:", resultado.shape)
```

Primero se verifica que el tamaño del vector sea B * C * H * W, porque esa es la cantidad total de datos necesarios para construir el tensor de imágenes. Si el vector tiene más o menos elementos, no se puede reorganizar correctamente.

Luego se aplica reshape(B, C, H, W). Esta operación no cambia los valores del vector, solo cambia la forma en que se agrupan. El resultado queda en formato NCHW, donde B representa el lote, C los canales, H la altura y W el ancho.

Después se utiliza transpose(tensor_nchw, (0, 2, 3, 1)). Esto cambia el orden de las dimensiones. La dimensión 0, que es B, queda igual. La dimensión 2, que es H, pasa al segundo lugar. La dimensión 3, que es W, pasa al tercer lugar. La dimensión 1, que es C, pasa al último lugar.

Por eso la forma cambia de:

(B, C, H, W)

a:

(B, H, W, C)

En conclusión, reshape organiza el vector como un tensor de cuatro dimensiones y transpose cambia el orden de los ejes. Los datos no se modifican; solamente se reorganiza su estructura.

### 5. Reduction Operations

```python
import numpy as np

def tensor_reductions(x, axis):
    """
    Aplica operaciones de reducción sobre un tensor.

    Calcula suma, media, máximo y argmax a lo largo de un eje específico.

    Parámetros:
        x: lista o arreglo con datos del tensor.
        axis: eje sobre el cual se aplican las reducciones.
              Puede ser un entero o None.

    Retorna:
        dict con las siguientes claves:
            "sum": suma de los elementos
            "mean": promedio de los elementos
            "max": valor máximo
            "argmax": índice del valor máximo
    """

    # 1. Convertir la entrada a arreglo NumPy
    tensor = np.array(x)

    # 2. Calcular la suma a lo largo del eje
    suma = np.sum(tensor, axis=axis)

    # 3. Calcular la media (promedio)
    media = np.mean(tensor, axis=axis)

    # 4. Encontrar el valor máximo
    maximo = np.max(tensor, axis=axis)

    # 5. Encontrar el índice del valor máximo
    indice_maximo = np.argmax(tensor, axis=axis)

    # 6. Guardar resultados en un diccionario
    resultados = {
        "sum": suma,
        "mean": media,
        "max": maximo,
        "argmax": indice_maximo
    }

    return resultados


# Prueba

x = [[1, 2, 3],
     [4, 5, 6]]

axis = 1

resultado = tensor_reductions(x, axis)

print("Suma:", resultado["sum"])
print("Media:", resultado["mean"])
print("Máximo:", resultado["max"])
print("Argmax:", resultado["argmax"])
```

Las operaciones de reducción permiten resumir la información de un tensor a lo largo de una dimensión específica.

El parámetro axis indica sobre qué eje se realiza la operación. Por ejemplo, si axis=1, se reduce cada fila, es decir, se combinan los valores de cada fila en un solo resultado.

Desde el punto de vista matemático:

- La suma calcula:
  sum = x₁ + x₂ + ... + xₙ

- La media calcula:
  mean = (1/n) * (x₁ + x₂ + ... + xₙ)

- El máximo selecciona el valor más grande del conjunto:
  max = max(x₁, x₂, ..., xₙ)

- El argmax devuelve la posición donde ocurre ese valor máximo.

Cuando se aplica una reducción, la dimensión correspondiente al eje se elimina. Por ejemplo, si el tensor tiene forma (2, 3) y se reduce en axis=1, el resultado tendrá forma (2,), ya que cada fila se convierte en un solo valor.

Estas operaciones no modifican los datos originales, sino que generan un resumen de menor dimensión, lo cual es fundamental en aprendizaje profundo para calcular pérdidas, estadísticas y decisiones de clasificación.


### 6. Vector Norms (L1, L2)

```python
import numpy as np
from typing import Dict

def compute_norms(x: np.ndarray) -> Dict[str, np.ndarray]:
   """
   Calcula las normas L1 y L2 para un lote de vectores.

   La norma L1 suma los valores absolutos de cada elemento del vector.
   La norma L2 calcula la raíz cuadrada de la suma de los cuadrados.
   Ambas producen un solo número por vector que representa su "tamaño".

   Variables
   ----------
   x : Matriz de entrada de forma (N, D), donde N es el número de vectores
       y D es la dimensión de cada uno.

   Retorna
   ----------
   Diccionario con:
     "l1" -> array de forma (N,) con la norma L1 de cada vector
     "l2" -> array de forma (N,) con la norma L2 de cada vector
   """

   # L1: sumar los valores absolutos a lo largo de cada fila (axis=1)
   # |x1| + |x2| + ... + |xD| para cada uno de los N vectores
   l1 = np.sum(np.abs(x), axis=1)

   # L2: raíz cuadrada de la suma de cuadrados por fila
   # sqrt(x1² + x2² + ... + xD²) para cada uno de los N vectores
   l2 = np.sqrt(np.sum(x**2, axis=1))

   return {"l1": l1, "l2": l2}


# Uso

x = np.array([
   [3,  4],   # Vector 0
   [1, -1]    # Vector 1
])

resultado = compute_norms(x)

print("L1:", resultado["l1"])  
print("L2:", resultado["l2"])
```

Dado un vector con varias componentes, una norma es el tamaño de un vector en un solo número.
La norma L1 suma los valores absolutos de cada componente.
La norma L2 es la distancia en línea recta desde el origen hasta el punto que representa el vector. Aplica basicamente el teorema de Pitágoras generalizado a muchas dimensiones.
Note que la norma L1 trata todos los errores de forma proporcional, mientras que la L2 castiga más fuertemente los errores grandes por el efecto del cuadrado.

### 7. Dot Product & Cross Product

```python
import numpy as np
from typing import Dict

def vector_products(a: np.ndarray, b: np.ndarray) -> Dict[str, np.ndarray]:
    """
    Calcula el producto punto y el producto cruz para lotes de vectores 3D.

    Parámetros:
        a: Shape (N, 3)
        b: Shape (N, 3)

    Retorna:
        Dict con "dot" (N,) y "cross" (N, 3)
    """

    # 1. Convertir entradas a arreglos NumPy preservando el dtype original
    a = np.asarray(a)
    b = np.asarray(b)

    # 2. Producto punto: multiplicar elemento a elemento y sumar por fila
    # Resultado shape (N,) → un escalar por par de vectores
    dot = np.sum(a * b, axis=1)

    # 3. Producto cruz: NumPy aplica la fórmula con precisión numérica correcta
    # Resultado shape (N, 3) → un vector 3D por par
    cross = np.cross(a, b)

    return {"dot": dot, "cross": cross}
```

Producto punto
Dados dos vectores a = (a_x, a_y, a_z) y b = (b_x, b_y, b_z):

a · b = a_x·b_x + a_y·b_y + a_z·b_z
Se multiplica cada componente con su par y se suman. El resultado es un escalar. Geométricamente esto equivale a:

a · b = ‖a‖ · ‖b‖ · cos(θ)
Si apuntan en la misma dirección → resultado positivo y grande
Si son perpendiculares → resultado cero
Si apuntan en direcciones opuestas → resultado negativo
Es por eso que el producto punto es una medida de similitud, y la base de los mecanismos de atención en Transformers (Q·Kᵀ).

En el código, np.sum(a * b, axis=1) multiplica elemento a elemento y colapsa las componentes en un solo número por fila.

Producto cruz
a × b = ( a_y·b_z − a_z·b_y,  a_z·b_x − a_x·b_z,  a_x·b_y − a_y·b_x )
A diferencia del punto, el resultado es un vector 3D con tres propiedades clave:

Es perpendicular a ambos vectores originales, por eso se usa para calcular normales de superficies en gráficos 3D.
Su magnitud es el área del paralelogramo formado por a y b (‖a × b‖ = ‖a‖·‖b‖·sin(θ)).
Es anticonmutativo: a × b = −(b × a), el orden importa.
Se usa np.cross en lugar de la fórmula manual porque con entradas float32 la implementación manual acumula errores de redondeo que superan la tolerancia de los tests.

### 8. Einstein Summation (einsum) Basics

```python
import numpy as np
from typing import Dict, Union

def einsum_ops(A: np.ndarray, B: np.ndarray) -> Dict[str, Union[np.ndarray, float]]:
    """
    Calcula operaciones básicas usando np.einsum.

    Parámetros:
        A: Shape (N, D)
        B: Shape (D, M)

    Retorna:
        Dict con "transpose", "sum", "row_sum", "col_sum", "matmul"
    """

    # 1. Convertir entradas a arreglos NumPy
    A = np.asarray(A)
    B = np.asarray(B)

    # 2. Transpuesta: intercambiar los índices i↔j
    # (N, D) → (D, N)
    transpose = np.einsum("ij->ji", A)

    # 3. Suma total: contraer todos los índices, sin índices de salida
    # Resultado: escalar
    total_sum = np.einsum("ij->", A)

    # 4. Suma por fila: conservar i, sumar sobre j
    # (N, D) → (N,)
    row_sum = np.einsum("ij->i", A)

    # 5. Suma por columna: conservar j, sumar sobre i
    # (N, D) → (D,)
    col_sum = np.einsum("ij->j", A)

    # 6. Multiplicación matricial: sumar sobre el índice compartido k
    # A(N,D) × B(D,M) → C(N,M)
    matmul = np.einsum("ik,kj->ij", A, B)

    return {
        "transpose": transpose,
        "sum": total_sum,
        "row_sum": row_sum,
        "col_sum": col_sum,
        "matmul": matmul,
    }
```

La notación de Einstein es una forma compacta de describir operaciones sobre tensores. La idea central es: los índices que aparecen en la entrada pero no en la salida se suman. Los que sí aparecen en la salida se conservan.

La notación sigue el patrón "indices_entrada->indices_salida".

Transpuesta — "ij->ji"
Una matriz A de shape (N, D) tiene dos índices: i para filas y j para columnas. Al escribir "ij->ji" simplemente se intercambian, convirtiendo filas en columnas y viceversa. El resultado es shape (D, N).

Suma total — "ij->"
No hay índices de salida. Eso significa que tanto i como j se contraen (se suman). El resultado es un único escalar: la suma de todos los elementos de A.

Suma por fila — "ij->i"
Se conserva i (filas) y se elimina j (columnas). Para cada fila i, se suman todos sus valores a lo largo de j:

row_sum[i] = Σⱼ A[i, j]
Resultado shape (N,).

Suma por columna — "ij->j"
Lo inverso: se conserva j y se elimina i. Para cada columna j, se suman todos sus valores a lo largo de i:

col_sum[j] = Σᵢ A[i, j]
Resultado shape (D,).

Multiplicación matricial — "ik,kj->ij"
Aquí intervienen dos matrices. A tiene índices ik (shape N×D) y B tiene índices kj (shape D×M). El índice k aparece en ambas entradas pero no en la salida, así que se suma sobre él:

C[i, j] = Σₖ A[i, k] · B[k, j]
Esto es exactamente la definición de multiplicación de matrices. El índice compartido k es la dimensión D que "conecta" ambas matrices, y contraerla produce el resultado de shape (N, M).

### 9. Advanced Einsum (Batch MatMul)

```python
import numpy as np

def batch_matmul(Q: np.ndarray, K: np.ndarray) -> np.ndarray:
    """
    Calcula Q @ K.T para atención multi-cabeza por lotes.

    Parámetros:
        Q: (B, H, S, D)
        K: (B, H, S, D)

    Retorna:
        Tensor (B, H, S, S)
    """

    # 1. Convertir entradas a arreglos NumPy
    Q = np.asarray(Q)
    K = np.asarray(K)

    # 2. Producto matricial por lote y por cabeza:
    # Para cada (b, h), calcular Q[b,h] @ K[b,h].T
    # Notación: "bhid,bhjd->bhij"
    #   - b: batch, h: cabeza, i: posición de query, j: posición de key
    #   - d: dimensión interna que se contrae (suma)
    scores = np.einsum("bhid,bhjd->bhij", Q, K)

    return scores
```

La idea es calcular, para cada secuencia del lote y cada cabeza de atención, el producto QK^T que mide qué tan similares son los queries y los keys.

La operación
Para cada batch b y cabeza h, tenemos dos matrices:

Q[b, h] de shape (S, D) → S queries, cada uno un vector de dimensión D
K[b, h] de shape (S, D) → S keys, cada uno un vector de dimensión D
El resultado es una matriz de scores de shape (S, S) donde:

Scores[b, h, i, j] = Σ_d Q[b, h, i, d] · K[b, h, j, d]
Es decir, el elemento (i, j) es el producto punto entre el query i-ésimo y el key j-ésimo. Geométricamente mide cuánto se "parecen" esos dos vectores: scores altos significan que el query i debe atender fuertemente al key j.

La cadena de einsum codifica exactamente esa fórmula:

bhid: índices del tensor Q (batch, head, posición de query, dimensión)
bhjd: índices del tensor K (batch, head, posición de key, dimensión)
->bhij: índices del resultado (batch, head, posición de query, posición de key)
El índice d aparece en ambas entradas pero no en la salida, así que se contrae (se suma sobre él). Eso es justamente el producto punto entre vectores de dimensión D.

Los índices b y h aparecen en todas partes sin contraerse: indican que la operación se hace independientemente para cada batch y cada cabeza, en paralelo.

Los índices i y j se mantienen libres porque queremos un score por cada par (query i, key j).

Resumen
Índice	Aparece en Q	Aparece en K	Aparece en salida	Qué hace einsum
b	    sí	            sí	            sí	                recorre en paralelo
h	    sí	        	sí		    	sí		    	    recorre en paralelo
i		sí		        no		        sí  	            recorre libre (filas del resultado)
j	    no		        sí		    	sí		    	    recorre libre (columnas del resultado)
d	    sí		        sí		        no		    	    se suma (contrae)
La d, que está en ambas entradas pero no en la salida, es la que convierte la operación en un producto punto.

### 10. Gradient of Sum

```python
import numpy as np
from typing import Tuple

def grad_sum(grad_y: float, x_shape: Tuple[int]) -> np.ndarray:
    """
    Dada la función y = sum(x), donde x es un vector,
    la derivada parcial de y respecto a x_i es 1 para cada i.
    Ahora, ∂L/∂x_i = (∂L/∂y)(∂y/∂x_i) = grad_y ⋅ 1 = grad_y.
    Esto quiere decir que cada entrada del vector grad_x es grad_y.
    
    Argumentos:
        grad_y: Gradiente escalar dL/dy
        x_shape: Forma del tensor x
        
    Retorna:
        Gradiente dL/dx con forma x_shape
    """
    
    return np.full(x_shape, grad_y)
```

### 11. Gradient of Matrix Multiplication

```python
import numpy as np
from typing import Dict

def grad_matmul(grad_C: np.ndarray, A: np.ndarray, B: np.ndarray) -> Dict[str, np.ndarray]:
    """
    Dada la operación de matrices C = AB,
    Se calcula ∂L/∂A y ∂L/∂B a partir de la igualdad C_{ij} = ∑_{k}^{K}A_{ik}B{kj}.
    
    Argumentos:
        grad_C: Gradiente upstream dL/dC (M, N)
        A: Input A (M, K)
        B: Input B (K, N)
        
    Retorna:
        Diccionario con "grad_A" y "grad_B"
    """
    
    grad_A = np.matmul(grad_C, np.transpose(B))
    grad_B = np.matmul(np.transpose(A), grad_C)
    
    return {"grad_A":grad_A, "grad_B":grad_B}
```

### 12. One-Hot Encoding

```python
import numpy as np

def one_hot_encode(indices: np.ndarray, num_classes: int) -> np.ndarray:
    """
    Dado un vector indices con N elementos y un número total de clases C,
    Se crea una matriz NxC donde cada fila tiene un 1 en la posición
    indicada en indices y 0 en el resto de posiciones.
    
    Argumentos:
        indices: Vector (N,)
        num_classes: Número total de clases C
        
    Retorna:
        Matriz one-hot (N, C)
    """

    N = len(indices)
    C = num_classes

    result = np.zeros((N, C))
    result[np.arange(N), indices] = 1

    return result
```

### 13. Softmax Implementation (Stable)

```python
import numpy as np

def softmax(x: np.ndarray) -> np.ndarray:
    """
    Convierte logits en probabilidades usando softmax numéricamente estable. Para cada fila se resta el valor máximo para evitar desbordamiento numérico sin cambiar el resultado matemático, ya que el máximo se cancela tanto en el numerador como en el denominador.

    Parámetros
    ----------
    x : matriz de forma (N, C)
        N muestras, cada una con C logits

    Retorna
    -------
    matriz de forma (N, C)
        Cada fila es una distribución de probabilidad que suma 1
    """

    # Restamos el máximo de cada fila para estabilidad numérica
    # keepdims=True mantiene la forma (N,1) para que el broadcasting funcione
    x_shifted = x - np.max(x, axis=1, keepdims=True)

    # Exponenciamos, entonces el mayor valor de cada fila será e^0 = 1
    exp_x = np.exp(x_shifted)

    # Dividimos cada fila entre su suma total para normalizar a probabilidades
    # keepdims=True en la suma permite la división fila a fila sin reshaping
    return exp_x / np.sum(exp_x, axis=1, keepdims=True)


# Uso
x = np.array([[2.0, 1.0, 0.1],
              [0.5, 2.5, 0.3]])

resultado = softmax(x)

print(resultado)

# Verificar que cada fila suma 1
print("Sumas por fila:", resultado.sum(axis=1)) 
```

El softmax convierte números arbitrarios (logits) en probabilidades y lo que hace es exponenciar cada valor para hacerlos positivos, y luego dividir cada uno entre la suma total para que todo sume 1. Si un logit es mayor que otro, su probabilidad resultante será mayor. 

### 14. Cross-Entropy Loss (Manual)

```python
import numpy as np

def cross_entropy_loss(probs, targets):
    """
    Calcula la pérdida de entropía cruzada promedio.

    Parámetros:
        probs: matriz (N, C) con probabilidades.
        targets: vector (N,) con índices de la clase correcta.

    Retorna:
        float: pérdida promedio.
    """

    # 1. Convertir entradas a arreglos NumPy
    probs = np.array(probs)
    targets = np.array(targets)

    # 2. Obtener tamaño del lote
    N = probs.shape[0]

    # 3. Definir épsilon para evitar log(0)
    eps = 1e-9

    # 4. Guardar las probabilidades correctas
    correct_probs = []

    # 5. Recorrer cada muestra y tomar la probabilidad de la clase correcta
    for i in range(N):
        correct_probs.append(probs[i][targets[i]])

    # 6. Convertir la lista a arreglo NumPy
    correct_probs = np.array(correct_probs)

    # 7. Calcular la pérdida promedio
    loss = -np.mean(np.log(correct_probs + eps))

    # 8. Convertir a float normal de Python
    return float(loss)

# Prueba

probs = [[0.1, 0.9],
         [0.8, 0.2]]

targets = [1, 0]

resultado = cross_entropy_loss(probs, targets)

print("Pérdida:", resultado)
```

La pérdida de entropía cruzada mide qué tan bien las probabilidades predichas por el modelo coinciden con la clase verdadera.

Primero, se tiene una matriz probs donde cada fila representa una distribución de probabilidad sobre las clases. Además, se tiene un vector targets que indica cuál es la clase correcta para cada muestra.

Para cada muestra i, se toma la probabilidad que el modelo asigna a la clase correcta, es decir:

p(i, y_i)

Esto se obtiene recorriendo cada fila del arreglo y seleccionando la posición correspondiente a la clase correcta usando un ciclo for.

Luego se aplica el logaritmo negativo:

-log(p(i, y_i))

Esto mide el error de la predicción:
- Si la probabilidad es alta (cerca de 1), la pérdida es pequeña.
- Si la probabilidad es baja (cerca de 0), la pérdida es grande.

Después, se promedian todas las pérdidas:

loss = -(1/N) * sum(log(p(i, y_i)))

donde N es la cantidad de muestras.

Se utiliza un pequeño valor epsilon (1e-9) para evitar problemas cuando la probabilidad es 0, ya que log(0) no está definido.

Finalmente, se devuelve un solo número que representa la pérdida promedio del modelo. Un valor pequeño indica buenas predicciones, mientras que un valor grande indica que el modelo está cometiendo errores.

### 15. Numerical Stability (Log-Sum-Exp)

```python
import numpy as np

def log_sum_exp(x: np.ndarray) -> np.ndarray:
   """
   Calcula log(suma(exp(x))) para cada fila.

   La identidad usada es:
       log(∑ eˣⁱ) = max(x) + log(∑ e^(xᵢ - max(x)))

   Parámetros
   ----------
   x : matriz de forma (N, D)
       N muestras, cada una con D valores.

   Retorna
   -------
   vector de forma (N,)
       log-sum-exp calculado independientemente para cada fila.
   """

   # máximo de cada fila
   m = np.max(x, axis=1, keepdims=True)

   # Restamos el máximo
   x_shifted = x - m
   # Exponenciar, sumar, aplicar log y sumar m
   # m.squeeze() elimina la dimensión extra para devolver forma (N,)
   return m.squeeze() + np.log(np.sum(np.exp(x_shifted), axis=1))


# Uso
x1 = np.array([[1000., 1001., 1002.]])
x2 = np.array([[2.0, 1.0, 0.1],
              [0.5, 2.5, 0.3]])

print(log_sum_exp(x1))  
print(log_sum_exp(x2))  
```

Queremos calcular log(∑ eˣⁱ), pero exponenciar números grandes produce infinito. La identidad algebraica permite factorizar e^max(x) dentro de la suma, y sacarlo como max(x) fuera del logaritmo, y lo que queda dentro del log son exponenciales de números negativos o cero. Se resta el máximo para que el exponencial más grande sea e⁰ = 1 y luego se suma al final para obtener el resultado final. Se puede ver como un cambio de escala midiendo desde el valor más grande. Todos los demás valores quedan en negativo, sus exponenciales quedan entre 0 y 1, y al final simplemente se suma de vuelta el máximo que se había apartado.
